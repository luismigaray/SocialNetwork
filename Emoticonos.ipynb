```python
import psycopg2
import sys, nltk, re, string
from psycopg2 import Error
from nltk.tokenize import TweetTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import word_tokenize
from nltk.data import load
from nltk.stem import SnowballStemmer
from string import punctuation
import demoji
import unidecode

from es_lemmatizer import lemmatize

import pandas as pd
import pandas.io.sql as sqlio

emoticonos_buenos = {'kissing face with smiling eyes', 'call me hand', 'winking face', 
                     'grinning squinting face', 'smiling face', 'hugging face', 'rolling on the floor laughing',
                     'sign of the horns', 'flexed biceps', 'kissing face', 'grinning face', 
                     'beaming face with smiling eyes', 'slightly smiling face', 'partying face', 'OK hand',
                     'smiling face with hearts', 'grinning face with big eyes', 'clapping hands',
                     'thumbs up', 'grinning face with sweat', 'smiling face with heart-eyes',
                     'smiling face with sunglasses', 'smiling face with halo', 'raising hands', 
                     'kissing face with closed eyes', 'love-you gesture', 'star-struck',
                     'smiling face with smiling eyes', 'grinning face with smiling eyes',
                     'face with tears of joy', 'victory hand'}

emoticonos_malos = {'frowning face with open mouth', 'angry face with horns', 'woozy face', 'tired face',
                    'fearful face', 'hot face', 'face with symbols on mouth', 'weary face', 'loudly crying face',
                    'disappointed face', 'face screaming in fear', 'pleading face', 'confused face', 
                    'expressionless face', 'pensive face', 'hushed face', 'face with steam from nose', 
                    'angry face', 'neutral face', 'dizzy face', 'confounded face', 'astonished face', 
                    'face with raised eyebrow', 'frowning face', 'exploding head', 'anguished face',
                    'persevering face', 'nauseated face', 'pouting face', 'face with open mouth', 
                    'downcast face with sweat', 'unamused face', 'face vomiting', 'thumbs down',
                    'worried face', 'face with monocle', 'crying face', 'zipper-mouth face', 'flushed face',
                    'sad but relieved face', 'slightly frowning face', 'anxious face with sweat', 'pile of poo'}

def crearConexion():
    
    connection = psycopg2.connect(user = "m.soto.montesinos",
                                      password = "1234",
                                      host = "127.0.0.1",
                                      port = "5432",
                                      database = "twitterBBVA")

    return connection

def obtenerTweets():

    try:
        
        connection = crearConexion()
        query = "select status_id, text, lang from training_set_bueno"
        
        tweets = sqlio.read_sql_query(query, connection)
        
    except (Exception, psycopg2.DatabaseError) as error :
    
        print ("Error obteniendo los datos", error)
    
    finally:
        
        if (connection):
            
            connection.close()
            print("PostgreSQL connection is closed")

    return tweets

def guardar_datos_db(tweet_id, esTweet, columna):

    try:
        
        connection = crearConexion()
        cursor = connection.cursor()
        
        if columna == 'bueno':
            
            if esTweet:
                query = """
                        UPDATE training_set_bueno SET emoticonos_buenos = 1 where status_id = '%s';
                        """ % (tweet_id)
            else:
                query = """
                        UPDATE training_set_bueno SET emoticonos_buenos = 0 where status_id = '%s';
                        """ % (tweet_id)
            
        elif columna == 'malo':
            
            if esTweet:
                query = """
                        UPDATE training_set_bueno SET emoticonos_malos = 1 where status_id = '%s';
                        """ % (tweet_id)
            else:
                query = """
                        UPDATE training_set_bueno SET emoticonos_malos = 0 where status_id = '%s';
                        """ % (tweet_id)
        
        cursor.execute(query)
        connection.commit()
        
    except (Exception, psycopg2.DatabaseError) as error :
    
        print ("Error obteniendo los datos", error)
    
    finally:
        
        if (connection):
            
            connection.close()
            cursor.close()
            print("PostgreSQL connection is closed")

def main():
    
    tweets = obtenerTweets()
    
    for index, row in tweets.iterrows():
    
        tweet = (row['text'])
        tweet_id = (row['status_id'])
        tweet_bueno_encontrado = False
        tweet_malo_encontrado = False
        
        tweet_emoticonos = demoji.findall(tweet).values()
        
        for w in emoticonos_buenos:
            
            if tweet_bueno_encontrado == False:
                
                for p in tweet_emoticonos:
                    if w in p:
                        tweet_bueno_encontrado = True
                        break
            else:
                break
                
        guardar_datos_db(tweet_id, tweet_bueno_encontrado, "bueno")
        
        for w in emoticonos_malos:
            
            if tweet_malo_encontrado == False:
                
                for p in tweet_emoticonos:
                    if w in p:
                        tweet_malo_encontrado = True
                        break
            else:
                break
                
        guardar_datos_db(tweet_id, tweet_malo_encontrado, "malo")
    
if __name__ == '__main__':
    main()





```
